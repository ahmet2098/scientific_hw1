Creating random A matrice and Toeplitz matrix by using the following references.
https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html
https://docs.scipy.org/doc/scipy/reference/generated/scipy.linalg.solve_toeplitz.html

```python
import numpy as np
from scipy.linalg import toeplitz

n, m = 200, 500
A = np.random.normal(loc=0, scale=1.0, size=(n, m))

randomArray = np.random.random(m)
B = toeplitz(randomArray, randomArray)
```

# 9.1)

```python
#Doing dot product operations by using dot() function.
#Transpose of the matrice can be used by adding .T
print("A + A")
print(A + A)

print("AxA.T")
result = A.dot(A.T)
print(result)

print("A.TxA")
result = A.T.dot(A)
print(result)

print("A.B")
print(np.dot(A, B))

print("A(B − λI)")
result = lambdaFun(A, B, 2.0)
print(result)

def lambdaFun(A, B, lamda):
    return np.dot(A, B-lamda*np.eye(m) )
```

# 9.2)

```python
# linalg.solve() is used to solve a linear matrix equationi or system of linear sclar equations
# https://numpy.org/doc/stable/reference/generated/numpy.linalg.solve.html
b = np.random.random(m)
x = np.linalg.solve(B, b)
print(x)
```

# 9.3)

```python
# We can create a type of matrix such as frobenius or infitiy by passing ord to the norm() function
# https://numpy.org/doc/stable/reference/generated/numpy.linalg.norm.html
frobeniusVector = np.linalg.norm(A,'fro')
print("Frobenius Norm:", frobeniusVector)

infitityVector = np.linalg.norm(B, np.inf)
print("Infinity Norm:", infitityVector)

largest = np.linalg.norm(B, 2)
smallest = np.linalg.norm(B, -2)
print("Largest singular:", largest)
print("Smallest singular:", smallest)
```

# 9.4)

```python
# Since time.clock() is deprecated, I have used perf_counter_ns instead to calculate time
from time import perf_counter_ns

t1_start = perf_counter_ns()
Z = np.random.standard_normal((n, n))
numOfIteration = 0
eigenValue = 0
eigenVector = np.ones(n)

while(True):
    normalized = np.dot(Z, eigenVector)

    eigenTemp = eigenValue
    eigenValue = np.linalg.norm(normalized)
    eigenVector = normalized/eigenValue
    if(abs(eigenTemp - eigenValue) <0.0003):
        break;
    numOfIteration += 1
t1_stop = perf_counter_ns()
print("the largest eigenvalue:",eigenValue)
print("the corresponding eigenvector")
print(eigenVector)
print("The number of iterations:", numOfIteration)
print("Computation Time:",t1_stop-t1_start, 'ns')
```
