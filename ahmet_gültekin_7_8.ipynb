## 7.2 Wordcount

### https://ocw.mit.edu/ans7870/6/6.006/s08/lecturenotes/files/t8.shakespeare.txt
### a) Find the 20 most common words

```python
def shakespeareWords():
    file = open('shakespeare.txt','r')
    file = file.read()
    dict = {}
    for word in file.lower().split():
        if word in dict:
            dict[word] = dict[word] + 1
        else:
            dict[word] = 1
    return dict
```

```python
import operator

words = shakespeareWords()

sortedWordsByValue = dict( sorted(words.items(), key=operator.itemgetter(1),reverse=True))

firstTwenty = list(sortedWordsByValue)[:20]
print(firstTwenty)
```

### b) How many unique words are used?

```python
file = open('shakespeare.txt','r')
file = file.read()

words = file.lower().split()
# Here because of the set property, I could eliminate the repeating words of each kind.
unique_words = set(words)
print(f'The number of words: {len(words)}')
print(f'The number of unique words: {len(unique_words)}')


```

### c) How many words are used at least 5 times?

```python
words = shakespeareWords()

result = 0
for word,count in words.items():
    if count >= 5:
        result += 1

print(f"There are {result} number of words used at least 5 times")
```

## 7.4 Sum of lists

### a) Data generation
### Write a function that takes three integers, n, a and b and a filename and writes to the file a list with n random integers between a and b.

```python
import random

def create(n,a,b,filename):
    f = open(filename, 'w+')
    for i in range(n):
        f.write(str(random.randint(a,b)) + " ")
        

create(20,1,50,'generation.txt')
create(20,1,50,'generation2.txt')
```

### b) Reading the data

```python
import string

def readFile(filename):
    f = open(filename,"r")
    line = f.read()
    return line
   
print(readFile('generation.txt'))
```

### c) Sum problem

```python
def find(filename1,filename2,k):
    u = []
    v = []
    list1 = readFile(filename1).split(" ")                                  
    list2 = readFile(filename2).split(" ")  
    for i in range(len(list1)-1):
        for j in range(len(list2)-1):
            if int(list1[i]) + int(list2[j]) == k:
                u.append(list1[i])
                v.append(list2[j])
    # set is used not to contain dublicate values
    u = set(u)
    v = set(v)
    return u,v

u,v = find('generation.txt','generation2.txt',32)
print(f'u: {u}')
print(f'v: {v}')
```

## 8.1 Rational numbers

```python
import math

class Rational:
    p = None
    q = None

    def __init__(self,p,q):
        self.p = p
        self.q = q

    def __str__(self):
        return f'{self.p}/{self.q}'

    def simplest(self):
        gcd = math.gcd(self.p,self.q)
        return f'{int(self.p / gcd)}/{int(self.q / gcd)}'

    def __add__(self,number,number2):
        return number + number2

    def __sub__(self,number,number2):
        return number - number2
    
    def __mul__(self,number,number2):
        return number * number2
    
    def __div__(self,number,number2):
        return number / number2
    
    def compare(self,number,number2):
        return number == number2
    
    def __float__(self,number):
        return number / 1.0


r1 = Rational(10,20)
print(r1)
print(r1.simplest())
print(r1.__add__(1/2,2/4))
print(r1.__sub__(1/2,2/4))
print(r1.__mul__(1/2,2/4))
print(r1.__div__(1/2,2/4))
print(r1.compare(1/2,2/4))
print(r1.__float__(1/2))
```

## 8.5 Set Class

```python
class mySet:
    """
    Class of Set that support add, issubset, issuperset
    union and intersection operations
    """
    def __init__(self):
        self.dict = {}

    def __iter__(self):
        return iter(self.dict.keys())

    def __contains__(self, item):
        return item in self.dict

    def add(self, item):
        self.dict[item] = 1

    def __len__(self):
        return len(self.dict)

    def issubset(self, t):
        key_set = self.dict.keys()
        for key in key_set:
            if key not in t:
                return False
        return True

    def issuperset(self, t):
        key_set = self.dict.keys()
        for key in t:
            if key not in key_set:
                return False
        return True

    def union(self, t):
        for key in t:
            self.dict[key] = 1

    def intersection(self, t):
        key_set = self.dict.keys()
        for key in key_set:
            if key not in t:
            	self.dict.pop(key)

    def copy(self):
        newSet = mySet()
        for e in self.dict():
            newSet.add(e)
        return newSet

    def __str__(self):
        string = '{'
        for e in self.dict.keys():
            string += str(e)
            string += ','
        string = string[:-1]
        return string + '}'
```
